```python
import pandas as pd
import matplotlib.pyplot as plt

plotdata = pd.DataFrame({

    "Objects opened with seed1":[5,4,8,8,10],

    "Objects opened with seed2":[5,7,6,9,20],

    "Objects opened with seed3":[4,6,6,8,25],
    
    "Total openable objects in the scene":[20,14,10,25,41]},

    index=["FloorPlan1", "FloorPlan3", "FloorPlan4", "FloorPlan5", "FloorPlan9"])

plotdata.plot(kind="bar",figsize=(15, 8))

plt.title("Number of opened objects in different environments")

plt.xlabel("Environment")

plt.ylabel("Number of objects")
```
